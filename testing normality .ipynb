# Testing of Normality of given data using R program

```r
#conda install -c r r-essentials 

ac <-read.csv("ac_freq.csv")
```

```r
head(ac)
```

```r
#Testing whether the data are normally distributed or not.
#Null hypothesis : the data is normally distributed
# Alternative Hypotheis : the data is not normally distributed.
```

```r
str(ac) #basic info
```

```r
shapiro.test(ac$frequency)
```

```r
#since the p value is greater than 0.05 therefore null hypotheris is accepted.
```

```r
#hist(ac$frequency, col = "red",xlab = "ac", ylab = "no of times", main = "Frequency histogram")
```

```r
x = ac$frequency 
x
```

```r
h = hist(x, col = "red", xlab = "ac" , ylab = "frequency")
xfit  = seq(min(x), max(x),length=10)
yfit = dnorm(xfit, mean = mean(x), sd = sd(x))
yfit = yfit*diff(h$mids[1:2])*length(x)
lines(xfit,yfit, col = "blue", lwd=2)
box()
```

```r
sd(x)
```

```r
se <- function(x) sqrt(var(x)/length(x))
se(x)
```

# For the distribution which  are normally distributed with unknown mean and known SD , a  (1 - alpha)100% CI isXbar +- z(alpha/2) * sigma/sqrt(n). We know n = 8, and are given sigma = 1.6142250.
# For a 95% CI, alpha = .05 and z(.025) 

```r
 me <- qnorm(.975)*(sd(x)/sqrt(8))  # margin of error 
me
```

```r
mean(x) - me 
```

```r
mean(x) + me
```

# Therefore the confidence interval is  (50.4814214306169, 52.7185785693831) i.e. the wide of the data variation.

# Data of monthly Irradiation at surkhet airport from 2005 to 2016 (A.D.)

```r

Rad <- read.csv("rad.csv")
R = Rad$Irradiation
H = hist(R, col = "green", xlab = "Irradiation", ylab = "Frequency")
xfit  = seq(min(R), max(R),length=10)
yfit = dnorm(xfit, mean = mean(R), sd = sd(R))
yfit = yfit*diff(H$mids[1:2])*length(R)
lines(xfit,yfit, col = "blue", lwd=2)
box() 
```

```r
shapiro.test(Rad$Irradiation)
```

```r
#Since p value is less than 0.05 , we can conclude that the data are not normally distributed.
```

 # QQ plot
 This plots the theoretical quantitle along x -axis and actual quantile along y axis.

```r
qqnorm(Rad$Irradiation)
qqline(Rad$Irradiation, col = 'red') 
```

# Since most of the points of Irradiation values does not lie along the line, we can conclude that the distribution is not normal.
